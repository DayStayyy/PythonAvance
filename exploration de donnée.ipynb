```python
#df['date'] = pd.to_datetime(df['date'], format='%Y%m%d%H%M%S')
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d %H:%M:%S')
# Create a new DataFrame with the mean temperature by year
mean_temp_by_year = df.groupby(df['date'].dt.year)['t'].mean().reset_index()
mean_temp_by_year.rename(columns={'t': 'TMean'}, inplace=True)
mean_temp_by_year
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d %H:%M:%S')
df['ff'] = pd.to_numeric(df['ff'], errors='coerce')
df = df.dropna(subset=['ff'])

# Group the DataFrame by week and calculate the range of temperatures within each week
df_weekly = df.groupby(df['date'].dt.isocalendar().week)['ff'].apply(lambda x: np.ptp(x)).reset_index(name='range_temp')
df_weekly
# Plot the temperature ranges within each week using a bar chart
plt.bar(df_weekly['week'], df_weekly['range_temp'])
plt.xlabel('Week')
plt.ylabel('Temperature Range')
plt.title('Temperature Range Within Each Week')
plt.show()
```
